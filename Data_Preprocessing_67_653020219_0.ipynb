* (10 points) จากข้อมูลที่ให้มา มีค่าว่างสูญหายมากน้อยแค่ไหน? และมีค่าผิดปกติ (outliers) และหรือข้อมูลรบกวน (noisy data) หรือไม่ในแต่ละตัวแปร?

```python
import matplotlib.pyplot as plt
# Check for missing values
missing_values = titanic.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Calculate the percentage of missing values
missing_percentage = (titanic.isnull().sum() / len(titanic)) * 100
print("\nPercentage of missing values in each column:\n", missing_percentage)

# Analyze outliers and noisy data (This is a general overview, further analysis may be needed)
# For numerical features, you can use box plots or histograms to visually identify outliers
# For categorical features, check for unexpected values or inconsistent data

# Example: Box plot for Age
plt.boxplot(titanic['age'].dropna())
plt.title('Box plot of Age')
plt.show()

# Example: Histogram for Fare
plt.hist(titanic['fare'].dropna(), bins=20)
plt.title('Histogram of Fare')
plt.show()

# Further analysis might include:
# - Identifying outliers using statistical methods (e.g., IQR, Z-score)
# - Exploring relationships between variables to detect inconsistencies
# - Investigating the nature of missing data (e.g., random missingness, non-random missingness)
```

* (10 points) จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

```python
# Fill missing values in 'age' with the median age
titanic['age'].fillna(titanic['age'].median(), inplace=True)

# Fill missing values in 'embarked' with the most frequent value
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)

# Fill missing values in 'fare' with the median fare
titanic['fare'].fillna(titanic['fare'].median(), inplace=True)

# Explanation:
# สำหรับ 'อายุ' การเติมด้วยค่ามัธยฐานเป็นตัวเลือกที่ดีเพราะมีความไวต่อค่าผิดปกติน้อยกว่าค่าเฉลี่ยและคิดว่าเป็น MCAR คือการหายไปที่เป็นสุ่มสมบูรณ์
# สำหรับ 'embarked' การเติมด้วยค่าที่พบมากที่สุดเป็นวิธีง่ายและมีประสิทธิภาพในการจัดการค่าที่หายไปในตัวแปรเชิงหมวดหมู่และคิดว่าเป็น MCAR แต่เนื่องจากไม่ใช่ตัวเลขก็เลยใช่เป็น mode
# สำหรับ 'ค่าโดยสาร' เช่นเดียวกับ 'อายุ' การเติมด้วยค่ามัธยฐานเป็นที่ต้องการเพื่อหลีกเลี่ยงอิทธิพลของค่าผิดปกติและคิดว่าเป็น MCAR คือการหายไปที่เป็นสุ่มสมบูรณ์

# Percentage of missing values ที่มีค่ามากๆติดไว้ก่อน ถ้าไม่จำเป็นยังไม่อยากลบหรือทำอย่างอื่นต่อเพราะกลัวข้อมูลสูญหายหรือเพิ่มโอกาศบิดเบือนข้อมูลมากขึ้นโดยไม่จำเป็น
```

```python
# Calculate the percentage of missing values
missing_percentage = (titanic.isnull().sum() / len(titanic)) * 100
print("\nPercentage of missing values in each column:\n", missing_percentage)
```

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
# i. Use one-hot encoding for the 'Embarked' variable
titanic = pd.get_dummies(titanic, columns=['embarked'])

# ii. Convert 'Sex' to numerical: male=0, female=1
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})

# iii. Create 'FamilySize'
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']

# iv. Create 'IsAlone'
titanic['IsAlone'] = np.where(titanic['FamilySize'] == 0, 1, 0)
```

```python
titanic
```

* ทำการ Normalize ข้อมูล \
    i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling \
    ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
    iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
# i. Min-Max Scaling for Fare and Age
titanic['fare'] = (titanic['fare'] - titanic['fare'].min()) / (titanic['fare'].max() - titanic['fare'].min())
titanic['age'] = (titanic['age'] - titanic['age'].min()) / (titanic['age'].max() - titanic['age'].min())

# ii. Create Fare_Per_Person
titanic['Fare_Per_Person'] = titanic['fare'] / (titanic['FamilySize'] + 1)

# iii. Create AgeGroup
titanic['AgeGroup'] = pd.cut(titanic['age'], bins=[0, 12/80, 18/80, 50/80, 1],)
```

* (10 points) นำเสนอการแจกแจงหรือทำความเข้าใจตัวแปรสำคัญในข้อมูล เช่น Age, Fare, and Pclass ด้วยการสร้างฮิสโตแกรมหรือแผนภาพกล่อง

```python
import matplotlib.pyplot as plt
# Histogram for Age
plt.hist(titanic['age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()

# Histogram for Fare
plt.hist(titanic['fare'], bins=20)
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Histogram of Fare')
plt.show()

# Box plot for Pclass
plt.boxplot([titanic[titanic['pclass'] == 1]['fare'],
              titanic[titanic['pclass'] == 2]['fare'],
              titanic[titanic['pclass'] == 3]['fare']])
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Box plot of Fare by Pclass')
plt.show()
```

* เมื่อเตรียมข้อมูลพร้อมแล้ว ให้หาความสัมพันธ์ของตัวแปรต่อไปนี้จากข้อมูล ซึ่งอาจนำเสนอด้วยสถิติและแผนภูมิหรือกราฟก็ได้: \
    i. (10 points) survived และ pclass \
    ii. (10 points) survived และ sex \
    iii. (10 points) survived และ age \
    iv. (10 points) จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare

```python
import matplotlib.pyplot as plt
import seaborn as sns

# i. survived และ pclass
print(titanic[['survived', 'pclass']].corr())
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.show()

# ii. survived และ sex
print(titanic[['survived', 'sex']].corr())
sns.countplot(x='sex', hue='survived', data=titanic)
plt.show()

# iii. survived และ age
print(titanic[['survived', 'age']].corr())
sns.boxplot(x='survived', y='age', data=titanic)
plt.show()

# iv. จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare
titanic['family_survived'] = titanic['survived'] * (titanic['FamilySize'] + 1)
print(titanic[['family_survived', 'fare']].corr())
sns.scatterplot(x='fare', y='family_survived', data=titanic)
plt.show()
```

* (10 points) สร้างตารางสรุปข้อมูลเพื่อหาคุณลักษณะของผู้โดยสารเรือไททานิค และศึกษาเกี่ยวกับการเสียชีวิตของผู้โดยสาร ทำการสรุปผลที่ได้จากตาราง

```python
titanic.pivot_table(index=['survived','pclass'], values=['age','sibsp','parch','fare'], aggfunc=['mean','count'])
```

จากตารางสรุปข้อมูล พบว่าผู้โดยสารที่เสียชีวิตส่วนใหญ่อยู่ใน pclass 3 มีอายุเฉลี่ย 26 ปี และมีค่าโดยสารเฉลี่ย 13.6 ส่วนผู้โดยสารที่รอดชีวิตส่วนใหญ่อยู่ใน pclass 1 มีอายุเฉลี่ย 35 ปี และมีค่าโดยสารเฉลี่ย 84.1

```python
titanic.pivot_table(index=['survived','sex'], values=['age','sibsp','parch','fare'], aggfunc=['mean','count'])
```
