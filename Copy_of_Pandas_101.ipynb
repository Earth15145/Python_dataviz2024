# HW 5
ใช้การ Query และ Data slicing เพื่อนตอบคำถามต่อไปนี้

1. นิทานที่ยาวที่สุด และสั้นที่สุด
2. ชนิดของนิทาน (type) ที่มากที่สุดและน้อยที่สุด
3. มี spatail ที่ไม่ใช่ไทยกี่เรื่อง

````
import pandas as pd
import re

# ฟังก์ชันลบตัวอักษรพิเศษ
def remove_special_characters(text):
    if isinstance(text, str): # Check if the input is a string
        return re.sub(r'\W+', '', text)  # \W+ matches any non-word characters
    else:
        return "" # Return an empty string if not a string

# โหลดข้อมูลจากไฟล์ CSV
file_path = 'path_to_your_file.csv'  # แทนที่ด้วยเส้นทางของไฟล์ CSV ของคุณ
folk_df = pd.read_csv(path_to_folktales)


def analyze_folk_tales(folk_df):
    # Remove special characters from descriptions and create a new column
    folk_df['description_clean'] = folk_df['description'].apply(remove_special_characters)

    # นิทานที่ยาวที่สุด และสั้นที่สุด (หลังจากลบตัวอักษรพิเศษ)
    longest_story = folk_df.loc[folk_df['description_clean'].str.len().idxmax()]
    shortest_story = folk_df.loc[folk_df['description_clean'].str.len().idxmin()]

    print(f"นิทานที่ยาวที่สุดคือ: {longest_story['title']} ({len(longest_story['description_clean'])} ตัวอักษร)")
    print(f"นิทานที่สั้นที่สุดคือ: {shortest_story['title']} ({len(shortest_story['description_clean'])} ตัวอักษร)")

    # ชนิดของนิทาน (type) ที่มากที่สุดและน้อยที่สุด
    type_counts = folk_df['type'].value_counts()
    most_common_type = type_counts.idxmax()
    least_common_type = type_counts.idxmin()

    print(f"ชนิดของนิทานที่มากที่สุด: {most_common_type} ({type_counts[most_common_type]} เรื่อง)")
    print(f"ชนิดของนิทานที่น้อยที่สุด: {least_common_type} ({type_counts[least_common_type]} เรื่อง)")

    # มี spatial ที่ไม่ใช่ไทยกี่เรื่อง
    non_thai_spatial_count = folk_df[folk_df['spatial'] != 'ไทย'].shape[0]

    print(f"มีนิทานที่ spatial ไม่ใช่ไทยทั้งหมด: {non_thai_spatial_count} เรื่อง")

# Call the function
analyze_folk_tales(folk_df)

````

# Quiz9

-วนลูปนับว่ามีนิทานที่มีคำว่าแม่หรือพ่อ ในชื่อเรื่อง มีกี่เรื่อง และเรื่องอะไรบ้าง

-มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน

-คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่ หรือ พ่อ มากที่สุด






````
mom_dad_count = 0
mom_life_count = 0
mom_local_count = 0
creator_counts = {}

for _, row in this_data.iterrows():
    title = row['title']
    story_type = row['type']
    creator = row['creator']
    if 'แม่' in title or 'พ่อ' in title:
        mom_dad_count += 1
        print(title)
        if story_type == 'นิทานชีวิต':
            mom_life_count += 1
        elif story_type == 'นิทานประจำถิ่น':
            mom_local_count += 1
        if creator in creator_counts:
            creator_counts[creator] += 1
        else:
            creator_counts[creator] = 1

print(f"มีนิทานที่มีคำว่าแม่หรือพ่อในชื่อเรื่อง {mom_dad_count} เรื่อง")
print("มีคำดังกล่าวในนิทานชีวิต" if mom_life_count > mom_local_count else "มีคำดังกล่าวในนิทานประจำถิ่น", "มากกว่ากัน")
print("คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อมากที่สุดคือ:", max(creator_counts, key=creator_counts.get))
````

````
# Convert 'Votes' and 'Rating' columns to numeric types
movie_df['Votes'] = pd.to_numeric(movie_df['Votes'], errors='coerce')
movie_df['Rating'] = pd.to_numeric(movie_df['Rating'], errors='coerce')

# Calculate the 'Score' column
movie_df['Score'] = (movie_df['Votes'] / movie_df['Votes'].max() * 50) + (movie_df['Rating'] / movie_df['Rating'].max() * 50)

# Sort the DataFrame by 'Score' in descending order
df = movie_df.sort_values(by='Score', ascending=False)

# Select the top 3 directors
top_directors = df.head(3)

# Display the result
top_directors
````
