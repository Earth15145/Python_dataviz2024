```python
file_path ='/content/drive/My Drive/Basic Python and Data Viz Midterm/test-TidyData_2024.xlsx'  # Assuming it's an Excel file
df = pd.read_excel(file_path)
df
```

```python
# Data Cleansing

# Handling missing values by dropping rows with null values in critical columns (e.g., 'Chlo_a', 'Chlo_b')
df_cleaned = df.dropna(subset=['Chlo_a', 'Chlo_b'])

# Display the cleaned data
df_cleaned
```

##1. à¸„à¸§à¸²à¸¡à¹à¸•à¸à¸•à¹ˆà¸²à¸‡à¸‚à¸­à¸‡ à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢ à¸‚à¸­à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ Chlo_a à¸—à¸¸à¸à¹† station à¸£à¸°à¸«à¸§à¹ˆà¸²à¸‡à¸›à¸µ 2018 à¹à¸¥à¸° à¸›à¸µ 2019

```python
# Extract year information from the 'month' column
df_cleaned['year'] = df_cleaned['month'].str.extract(r'(\d{4})').astype(int)

# Filter data for years 2018 and 2019
df_2018 = df_cleaned[df_cleaned['year'] == 2018]
df_2019 = df_cleaned[df_cleaned['year'] == 2019]

# Calculate the average Chlo_a for each station for 2018 and 2019
avg_chlo_a_2018 = df_2018.groupby('station')['Chlo_a'].mean().reset_index(name='avg_Chlo_a_2018')
avg_chlo_a_2019 = df_2019.groupby('station')['Chlo_a'].mean().reset_index(name='avg_Chlo_a_2019')

# Merge the two dataframes to find the difference
chlo_a_diff = pd.merge(avg_chlo_a_2018, avg_chlo_a_2019, on='station')
chlo_a_diff['difference'] = chlo_a_diff['avg_Chlo_a_2019'] - chlo_a_diff['avg_Chlo_a_2018']

# Display the differences
chlo_a_diff
```

## 2. à¸ªà¸£à¹‰à¸²à¸‡à¸•à¸²à¸£à¸²à¸‡à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢ Chlo_b à¸—à¸¸à¸à¹† station à¸‚à¸­à¸‡à¹à¸•à¹ˆà¸¥à¸°à¹€à¸”à¸·à¸­à¸™

```python
# Calculate the average Chlo_b for each station in each month
avg_chlo_b_by_month = df_cleaned.groupby(['station', 'month'])['Chlo_b'].mean().reset_index()

# Display the table of average Chlo_b values
avg_chlo_b_by_month
```

```python
# Create a Pivot Table for the average Chlo_b
chlo_b_pivot = avg_chlo_b_by_month.pivot(index='station', columns='month', values='Chlo_b')

# Display the Pivot Table
chlo_b_pivot
```
